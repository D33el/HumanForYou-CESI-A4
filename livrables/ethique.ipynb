# **Contexte**

L'entreprise de produits pharmaceutiques HumanForYou basée en Inde emploie environ 4000 personnes. Cependant, chaque année elle subit un turn-over d'environ 15% de ses employés nécessitant de retrouver des profils similaires sur le marché de l'emploi.

La direction trouve que ce niveau de turn-over n'est pas bon pour l'entreprise car :

* Les projets sur lesquels étaient les employés quittant la société prennent du retard ce qui nuit à la réputation de l'entreprise auprès de ses clients et partenaires.
* Un service de ressources humaines de taille conséquente doit être conservé car il faut avoir les moyens de trouver les nouvelles recrues.
* Du temps est perdu à l'arrivée des nouveaux employés car ils doivent très souvent être formés et ont besoin de temps pour devenir pleinement opérationnels dans leur nouvel environnement.

Le direction souhaite déterminer les facteurs ayant le plus d'influence sur ce taux de turn-over pour définir des pistes d'amélioration pour donner à ses employés l'envie de rester.

## **Conception**
Avant de commencer le développement (ou la construction) de notre intelligence artificielle, il faut déterminer quels sont les caractéristiques pouvant être utilisés tout en respectant les notions d'éthique suivantes : 

- **La liberté humaine :** Un être humain est libre de faire ses propres choix.
- **La dignité humaine :** Un être humain doit être considéré comme une fin en soi et non comme un moyen, c'est-à-dire que l'on ne peut pas rabattre un être humain à un simple objet.
- **L'autonomie humaine :** Elle est définie comme la capacité du sujet à décider pour lui-même des règles auxquelles il obéit, et à agir en conséquence.
- **La protection de la vie privée et des données :** vise à réglementer l’utilisation et le traitement des données personnelles que nous respectons via la confidentialité des données assuré par :
    - <u>L’anonymat :</u> Aucun employé n’est nommé ou ne peut être rattaché à notre jeu de données. 
    - <u>La synthétisation :</u> Nous avons recréé des données à partir de la réalité. 
- **La non-discrimination et l'égalité :** Principe interdisant de traiter moins favorablement une personne en se basant sur des critères réels ou supposés comme l'apparence, la croyance, l'âge ou le sexe pour se rapprocher d’une égalité de traitement.
- **La diversité :** hétérogénéité
- **L'équité :** Principe impliquant le respect absolu de ce qui est dû à chacun ou de l’égalité des chances. Il vise à instaurer une égalité de droit, en tenant compte des inégalités de fait.

## **Construction**

Avoir des données éthiques ne suffit pas à garantir que notre IA le soit. En effet, la conception d’un modèle d’intelligence artificielle doit répondre à des exigences strictes en matière de **sécurité** et de **robustesse**.  

La robustesse et la sécurité d’une IA se mesurent à sa capacité à s’adapter à une grande diversité de cas d’utilisation. Un modèle doit être entraîné sur un volume de données conséquent, mais la qualité de ces données est tout aussi essentielle que leur quantité. Si l’IA est uniquement exposée à des situations idéales, elle risque de ne savoir gérer que des scénarios simples et d’échouer face à des cas plus complexes.  

Prenons l’exemple de l’apprentissage de la conduite. Rouler sur une route déserte est relativement simple, mais circuler dans une métropole comme Paris, avec un trafic dense et de nombreux imprévus, est un tout autre défi. De la même manière, une IA doit être exposée à une variété de situations pour développer des compétences véritablement efficaces.  

De plus, il est crucial de prendre en compte le **sur-apprentissage**. En intelligence artificielle, le sur-apprentissage survient lorsqu’un modèle mémorise excessivement les spécificités des données d’entraînement, au point de perdre sa capacité à généraliser à de nouveaux cas. Cela peut conduire à des performances impressionnantes sur l’ensemble d’entraînement, mais médiocres en conditions réelles.  

L’illustration suivante illustre ce phénomène : le modèle de gauche montre une généralisation efficace, tandis que celui de droite est trop ajusté aux données d’apprentissage, compromettant ainsi sa fiabilité en dehors de cet échantillon.

## **Utilisation**  

L’intelligence artificielle est souvent perçue de manière biaisée en raison de son hypermédiatisation, des représentations erronées et des fictions futuristes qui en altèrent la compréhension. Cette méconnaissance alimente une certaine méfiance quant à son utilisation et à ses implications. Pour pallier ce problème, il est essentiel d’adopter une approche pédagogique et transparente, en exposant clairement son fonctionnement.  

Cela implique une description détaillée et accessible de chaque phase du cycle de vie de l’IA, selon les principes suivants :  

- **Clarté du processus** : Chaque étape doit être explicitée de manière compréhensible afin que les utilisateurs puissent saisir avec précision les opérations effectuées, de la collecte des données à la prise de décision.  
- **Interprétabilité des résultats** : Les sorties du modèle doivent être formulées de manière à être facilement analysables et exploitables. L’utilisateur doit pouvoir en contester la validité sur la base d’informations accessibles, mettant en lumière les facteurs et la logique sous-jacents aux prédictions générées.  

Une transparence accrue favorisera ainsi une meilleure acceptation et une utilisation plus éclairée de l’intelligence artificielle.
